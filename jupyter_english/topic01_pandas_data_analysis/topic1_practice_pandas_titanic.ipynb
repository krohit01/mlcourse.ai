**Let's create a new feature.**

```python
def age_category(age):
    '''
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    '''
    if age < 30:
        return 1
    elif age < 55:
        return 2
    elif age >= 55:
        return 3
```

```python
age_categories = [age_category(age) for age in data.Age]
data['Age_category'] = age_categories
```

**Another way is to do it with `apply`.**

```python
data['Age_category'] = data['Age'].apply(age_category)
```

```python
## split name into 2 cols
data[['Last_Name','First_Name']] = data.Name.str.split(", ",expand=True) 

## clean name of prefixes
def clean_name(name):
    name=name.replace("Mr.","")
    name=name.replace("Mrs.","")
    name=name.replace("Miss.","")
    return name

data['First_Name']=data['First_Name'].apply(clean_name)
data.head()
```
